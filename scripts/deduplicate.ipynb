```python
import argparse
import os

import pandas as pd


def dedup_by_title(input_path: str, output_path: str) -> None:
    df = pd.read_csv(input_path)
    title_col = next((c for c in df.columns if c.lower() == "title"), None)
    if not title_col:
        raise KeyError("Input CSV missing a Title column")

    title_norm = df[title_col].astype(str).str.strip().str.lower()
    #remove punctuation from title_norm
    title_norm = title_norm.str.replace(r'[^\w\s]', '', regex=True)
    deduped = df.loc[~title_norm.duplicated()].copy()
    deduped.to_csv(output_path, index=False)

input_path = "../scripts/outputs/labeled_clusters_subset_2.csv"
output_path = "../scripts/outputs/labeled_clusters_subset_2.csv"
dedup_by_title(input_path, output_path)

```
